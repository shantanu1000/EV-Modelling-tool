```python
import matplotlib.pyplot as plt
import random

bus_data = {
    "Blue Bird Vision Electric": {"type": "C", "capacity": 155, "range": 120},
    "Collins Bus Corporation and Motiv-Power Systems Type A Electric School Bus": {"type": "A", "capacity": 127, "range": 105},
    "Daimler’s Thomas Built Buses and Proterra Saf-T-Liner C2 Jouley series": {"type": "C", "capacity": 220, "range": 135},
    "GreenPower Synapse 72 fully electric": {"type": "D", "capacity": 200, "range": 140},
    "Lion Electric LionC": {"type": "C", "capacity": 220, "range": 155},
    "Trans Tech and Motiv Power Systems SST-e": {"type": "A", "capacity": 106, "range": 85},
}

charger_data = {
    "DCFC - 50kW": 50,
    "DCFC - 100kW": 100,
    "DCFC - 120kW": 120,
    "DCFC - 160kW": 160,
    "DCFC - 175kW": 175,
    "DCFC - 500kW": 500,
}

def display_bus_info(selected_buses, charger_type, num_chargers):
    total_buses = sum(count for bus_type, count in selected_buses.items())
    total_capacity = sum(count * bus_data[bus_type]['capacity'] for bus_type, count in selected_buses.items())
    
    print("\nSelected Buses:")
    for bus_type, count in selected_buses.items():
        print(f"{bus_type}: {count} buses")
    print(f"Total number of buses: {total_buses}")
    print(f"Total energy storage capacity: {total_capacity} kWh")
    
    print(f"\nSelected Charger Type: {charger_type}")
    print(f"Charger Capacity: {charger_data[charger_type]} kW")
    total_charger_capacity = num_chargers * charger_data[charger_type]
    print(f"Total charging capacity for {num_chargers} chargers: {total_charger_capacity} kW")

class Bus:
    def __init__(self, capacity):
        self.capacity = capacity  # kWh
        self.initial_charge = self.capacity * random.uniform(0.15, 0.4)  # kWh
        self.needed_charge = self.capacity - self.initial_charge  # kWh

def cheapest_load_and_buses(buses, charging_window, charger_power, electricity_rates, simultaneous_buses):
    buses = sorted(buses, key=lambda x: x.needed_charge, reverse=True)
    total_cost = 0

    # List to store how many buses are charged each hour and charge delivered each hour
    buses_charged_each_hour = [0] * charging_window
    charge_delivered_each_hour = [0] * charging_window

    for bus in buses:
        remaining_charge = bus.needed_charge
        for hour, rate in sorted(enumerate(electricity_rates), key=lambda x: x[1]):
            if remaining_charge <= 0:
                break
            if charger_power * (buses_charged_each_hour[hour] + 1) <= charger_power * simultaneous_buses:
                charge_this_hour = min(remaining_charge, charger_power)
                total_cost += charge_this_hour * rate
                remaining_charge -= charge_this_hour
                buses_charged_each_hour[hour] += 1
                charge_delivered_each_hour[hour] += charge_this_hour

    return total_cost, buses_charged_each_hour, charge_delivered_each_hour


def main():
    print("Welcome to the Energy Modeling Tool\n")
    selected_buses = {}
    
    while True:
        # Bus Selection
        print("Available bus types:")
        for i, bus_type in enumerate(bus_data.keys()):
            print(f"{i+1}. {bus_type}")
        while True:
            try:
                bus_type_index = int(input("\nSelect the type of bus by entering its number (or enter 0 to finish bus selection): ")) - 1
                if bus_type_index == -1:
                    break  # Finish bus selection
                if bus_type_index < 0 or bus_type_index >= len(bus_data):
                    raise ValueError
                break
            except ValueError:
                print("Invalid input. Please enter a number corresponding to a bus type.")
        if bus_type_index == -1:
            break  # Finish bus selection
        bus_type = list(bus_data.keys())[bus_type_index]
        while True:
            try:
                num_buses = int(input(f"\nEnter the number of {bus_type} buses: "))
                if num_buses <= 0:
                    raise ValueError
                break
            except ValueError:
                print("Invalid input. Please enter a positive integer.")
        selected_buses[bus_type] = num_buses
    
    # Charger Selection
    print("\nAvailable charger types:")
    for i, charger_type in enumerate(charger_data.keys()):
        print(f"{i+1}. {charger_type}")
    while True:
        try:
            charger_type_index = int(input("\nSelect the type of charger by entering its number: ")) - 1
            if charger_type_index < 0 or charger_type_index >= len(charger_data):
                raise ValueError
            break
        except ValueError:
            print("Invalid input. Please enter a number corresponding to a charger type.")
    charger_type = list(charger_data.keys())[charger_type_index]
    while True:
        try:
            num_chargers = int(input("\nEnter the number of chargers: "))
            if num_chargers <= 0:
                raise ValueError
            break
        except ValueError:
            print("Invalid input. Please enter a positive integer.")
    while True:
        try:
            charging_window = int(input("\nEnter the charging window (in hours): "))
            if charging_window <= 0:
                raise ValueError
            break
        except ValueError:
            print("Invalid input. Please enter a positive integer.")
        
    # Electricity Rates Input
    electricity_rates = []
    print("\nPlease enter the electricity rate for each hour of the charging window:")
    for i in range(charging_window):
        while True:
            try:
                rate = float(input(f"Hour {i+1} ($/kWh): "))
                if rate < 0:
                    raise ValueError
                electricity_rates.append(rate)
                break
            except ValueError:
                print("Invalid input. Please enter a non-negative number.")
    
    # Calculating total capacity needed and charger capacity
    total_needed_charge = sum(count * bus_data[bus_type]['capacity'] for bus_type, count in selected_buses.items())
    charger_capacity = charger_data[charger_type]
    
    
    
    buses = []
    for bus_type, count in selected_buses.items():
        for _ in range(count):
            buses.append(Bus(bus_data[bus_type]['capacity']))
    
    # Determine simultaneous_buses using the logic from your first code
    total_needed_charge = sum(bus.needed_charge for bus in buses)
    simultaneous_buses = total_needed_charge / (charging_window * charger_capacity)
    
    # Call the cheapest_load_and_buses function
    total_cost, buses_charged_each_hour, charge_delivered_each_hour = cheapest_load_and_buses(
        buses, charging_window, charger_capacity, electricity_rates, simultaneous_buses
    )
    # Displaying the information
    print(f"\nTotal Needed Charge: {total_needed_charge:.2f} kWh")
    print(f"Total Cost: ${total_cost:.2f}")
    
    
    print(f"The total energy needed to charge all buses is {total_needed_charge:.2f} kWh")
    print(f"The total cost to charge all buses is ${total_cost:.2f}")
    print("Hourly Details:")
    for hour in range(charging_window):
        print(f" - Hour {hour + 1}:")
        print(f"    Buses Charged: {buses_charged_each_hour[hour]}")
        print(f"    Charge Delivered: {charge_delivered_each_hour[hour]:.2f} kWh")
        print(f"    Electricity Rate: ${electricity_rates[hour]:.2f}/kWh")
        
    # Displaying bus and charger selection
    display_bus_info(selected_buses, charger_type, num_chargers)
    
    plt.plot(range(charging_window), charge_delivered_each_hour, marker='o')
    plt.xticks(range(charging_window), [f"Hour {i+1}" for i in range(charging_window)])
    plt.xlabel("Charging Window")
    plt.ylabel("Charge Distributed (kWh)")
    plt.title("Charge Distribution Across the Charging Window")
    plt.grid(True)
    plt.show()
if __name__ == "__main__":
    main()
```
